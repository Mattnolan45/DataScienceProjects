```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("IrelandDataCleaned.csv")
print(data.head())
```

```python
plt.plot(data['Unnamed: 0'][::-1],data['cases'], label="Cases")
plt.plot(data['Unnamed: 0'][::-1],data["deaths"], label="Deaths")
plt.title('Covid Progression Ireland')
plt.xlabel('Day')
plt.ylabel('Daily Cases')
plt.legend()
plt.show()
```
